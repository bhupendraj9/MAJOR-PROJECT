```python
import matplotlib.pyplot as plt
from PIL import Image

plot_paths = [
    'runs/detect/road_sign_train/confusion_matrix.png',
    'runs/detect/road_sign_train/PR_curve.png',
    'runs/detect/road_sign_train/F1_curve.png',
    'runs/detect/road_sign_train/results.png'
]

titles = [
    'Confusion Matrix',
    'Precision-Recall Curve',
    'F1 Curve',
    'Training Results (Loss/Metric Curves)'
]

for path, title in zip(plot_paths, titles):
    try:
        img = Image.open(path)
        plt.figure(figsize=(10, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title(title)
        plt.show()
    except Exception as e:
        print(f"Could not display {title}: {e}")
```

```python
import yaml
import os
import matplotlib.pyplot as plt

with open('/content/data/data.yaml', 'r') as f:
    data = yaml.safe_load(f)
class_names = data['names']

label_dir = '/content/data/train/labels'
labels = []
for label_file in os.listdir(label_dir):
    with open(os.path.join(label_dir, label_file)) as lf:
        for line in lf:
            class_id = int(line.split()[0])
            labels.append(class_names[class_id])

plt.figure(figsize=(12, 5))
plt.hist(labels, bins=len(class_names), rwidth=0.8)
plt.xticks(rotation=90)
plt.title('Class Distribution in Training Set')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
